## 4. Exploratory Data Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## a. Univariate Analysis (Single feature)

```python
# Distribution of Layoff Risk -- target variable

plt.figure(figsize=(8, 5))

sns.countplot(
    data=dataset, 
    x='Layoff_Risk',
    order=['Low', 'Medium', 'High'],
    palette ='tab20')

plt.title('Distribution of Layoff Risk')
plt.xlabel('Layoff Risk')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution

plt.figure(figsize=(10,5))

sns.histplot(
    dataset["Age"],
    bins=20,
    kde=True,
    color="steelblue" )

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Education Level Distribution

plt.figure(figsize=(10,5))

sns.countplot(
    data=dataset,
    x="Education_Level",
    hue="Education_Level",
    palette="viridis",
    legend=False )

plt.xticks(rotation=45)
plt.title("Education Level Distribution")
plt.show()
```

```python
# Experience Distribution

plt.figure(figsize=(10,5))

sns.histplot(
    dataset["Years_of_Experience"],
    bins=30,
    kde=True,
    color="purple"
)

plt.title("Years of Experience Distribution")
plt.show()
```

```python
# Industry Distribution

plt.figure(figsize=(10,5))

sns.countplot(
    data=dataset,
    x="Industry",
    hue="Industry",
    palette="viridis",
    legend=False )

plt.xticks(rotation=45)
plt.title("Industry Distribution")
plt.show()
```

```python
# Job Role Distribution

plt.figure(figsize=(15,5))

sns.countplot(
    data=dataset,
    x="Job_Role",
    hue="Job_Role",
    palette="viridis",
    legend=False )

plt.xticks(rotation=45)
plt.title("Job Role Distribution")
plt.show()
```

## b. Feature vs Target Analysis

```python
# Age vs layoff risk

plt.figure(figsize=(9, 5))

sns.boxplot(
    data=dataset,
    x='Layoff_Risk', 
    y='Age', 
    order=['Low', 'Medium', 'High'], 
    palette='tab20' )

plt.title('Age Distribution vs Layoff Risk Levels')
plt.xlabel('Layoff_Risk')
plt.ylabel('Age')
plt.show()
```

```python
# Experience vs layoff risk

plt.figure(figsize=(9, 5))

sns.violinplot(
    data=dataset,
    x='Layoff_Risk',
    y='Years_of_Experience', 
    order=['Low', 'Medium', 'High'],
    palette='Reds')

plt.title('Experience vs Layoff Risk Levels')
plt.xlabel('Layoff_Risk')
plt.ylabel('Years_of_Experience')
plt.show()
```

```python
# Education Level vs layoff risk

plt.figure(figsize=(9, 5))

sns.boxplot(
    data=dataset,
    x='Layoff_Risk',
    y='Education_Level', 
    order=['Low', 'Medium', 'High'],
    palette='bright')

plt.title('Education Level vs Layoff Risk Levels')
plt.xlabel('Layoff_Risk')
plt.ylabel('Education_Level')
plt.show()
```

```python
# Routine Task Percentage vs Layoff Risk

plt.figure(figsize=(9, 5))

sns.violinplot(
    data=dataset,
    x='Layoff_Risk',
    y='Routine_Task_Percentage', 
    order=['Low', 'Medium', 'High'] )

plt.title('Routine Task Percentage vs Layoff Risk')
plt.xlabel('Layoff_Risk')
plt.ylabel('Routine_Task_Percentage')
plt.show()
```

```python
# Layoff Risk Across Industries

plt.figure(figsize=(12, 5))

sns.countplot(
    data=dataset,
    x='Industry',
    hue='Layoff_Risk',
    hue_order=['Low', 'Medium', 'High'],
    palette=['#4CAF50', '#FF9800', '#F44336'])

plt.title('Layoff Risk Across Industries')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.legend(title='Layoff Risk')

plt.xticks(rotation=30)
plt.show()
```

```python
# Tasks Automated Percentage VS Layoff Risk

plt.figure(figsize=(10,6))

sns.boxplot(
    data=dataset,
    x="Layoff_Risk",
    y="Tasks_Automated_Percentage",
    hue="Layoff_Risk",
    palette="cool",
    legend=False
)

plt.title("Automation Percentage vs Layoff Risk")
plt.show()
```

## c. Feature to Feature Analysis

```python
# Correlation Heatmap

numeric_df=dataset.select_dtypes(include=np.number)
plt.figure(figsize=(14,9))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    linewidth=0.5 )

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Top 15 Job Roles with High Layoff Risk 

filtered = dataset[dataset['Layoff_Risk'] == 'High']

top_roles = filtered['Job_Role'].value_counts().head(15).index

plt.figure(figsize=(12,7))

sns.countplot(
    data=filtered[filtered['Job_Role'].isin(top_roles)],
    y='Job_Role',
    order=top_roles,
    color = 'steelblue'
)

plt.title('Top 15 Job Roles with High Layoff Risk')
plt.xlabel('Count')
plt.ylabel('Job Role')

plt.tight_layout()
plt.show()
```

```python
# Experience vs Age

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=dataset, 
    x="Years_of_Experience",
    y="Age", 
    hue="Layoff_Risk", 
    palette="bright")

plt.title("Experience vs Age")
plt.show()
```

```python
#Top 5 Job Roles by Average Creativity Requirement

top_roles = (
    dataset.groupby('Job_Role')['Creativity_Requirement']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure(figsize=(5,5))

sns.barplot(
    x=top_roles.index,
    y=top_roles.values
)

plt.title('Top 5 Job Roles by Average Creativity Requirement')
plt.xlabel('Job Role')
plt.ylabel('Average Creativity Requirement')

plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
# Routine Task Percentage vs Tasks Automated Percentage

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=dataset, 
    x="Routine_Task_Percentage",
    y="Tasks_Automated_Percentage", 
    hue="Layoff_Risk", 
    palette="cool")

plt.title("Routine_Task_Percentage vs Tasks_Automated_Percentage")
plt.show()
```

##### 3. Years of experience show varying relationships with risk, indicating that experience alone does not guarantee job security.
